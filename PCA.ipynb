```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data = pd.read_csv('D:/Data Science/Assignment/08PCA/wine.csv')
data.head()
```

```python
data.describe()
```

```python
norm=scale(data)
norm
```

```python
df=pd.DataFrame(norm,columns=data.columns)
df
```

```python
pca=PCA()
pca_values=pca.fit_transform(norm)
```

```python
pca_values.shape
```

```python
pca=PCA(n_components=3)
pca_values=pca.fit_transform(norm)
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1,color='red')
```

```python
pca_values[:,0:1]
```

```python
pca_df=pd.DataFrame(pca_values,columns=['pca1','pca2','pca3'])
pca_df
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
wcss=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,random_state=0).fit(pca_df)
    wcss.append(kmeans.inertia_)
    print(kmeans)
    print(kmeans.inertia_)
    print(kmeans.labels_)
```

```python
sns.pointplot(list(range(1,10)),wcss)
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendogram=sch.dendrogram(sch.linkage(pca_df,method='complete'))
plt.show()
```

```python
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete')
y_hc=hc.fit_predict(pca_df)
```

```python
y_hc
```

```python
pca_df['label']=pd.Series(y_hc)
pca_df
```
